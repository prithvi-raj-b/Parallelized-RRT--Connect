```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib widget
```

```python
plt.figure()
with open("map1.dat") as f:
    _ = f.readline()
    lines = [[int(z) for z in y if z!=''] for y in [x.split(' ') for x in f.read().split('\n')] if y!=['']]
    plt.imshow(lines, cmap='gray_r', origin='lower',extent=[-5.2,4.8,-5.2,4.8])
with open("graph_start.txt") as f1:
    lines = f1.read().split('\n')
    xpath = [ (float(x[0]),float(x[2])) for x in [y.split(' ') for y in lines] if x!=['']]
    ypath = [ (float(x[1]),float(x[3])) for x in [y.split(' ') for y in lines] if x!=['']]
    for i in range(len(xpath)):
        plt.plot(xpath[i], ypath[i], 'r-')
with open("graph_finish.txt") as f1:
    lines = f1.read().split('\n')
    xpath = [ (float(x[0]),float(x[2])) for x in [y.split(' ') for y in lines] if x!=['']]
    ypath = [ (float(x[1]),float(x[3])) for x in [y.split(' ') for y in lines] if x!=['']]
    for i in range(len(xpath)):
        plt.plot(xpath[i], ypath[i], 'b-')
```

```python
plt.figure()
with open("map1.dat") as f:
    _ = f.readline()
    lines = [[int(z) for z in y if z!=''] for y in [x.split(' ') for x in f.read().split('\n')] if y!=['']]
    plt.imshow(lines, cmap='gray_r', origin='lower',extent=[-5.2,4.8,-5.2,4.8])
with open("nodes0.dat") as f1:
    lines = f1.read().split('\n')
    xpath = [ (float(x[0]),float(x[2])) for x in [y.split(' ') for y in lines] if x!=['']]
    ypath = [ (float(x[1]),float(x[3])) for x in [y.split(' ') for y in lines] if x!=['']]
    for i in range(len(xpath)):
        plt.plot(xpath[i], ypath[i], '-', color='orange')
with open("nodes1.dat") as f1:
    lines = f1.read().split('\n')
    xpath = [ (float(x[0]),float(x[2])) for x in [y.split(' ') for y in lines] if x!=['']]
    ypath = [ (float(x[1]),float(x[3])) for x in [y.split(' ') for y in lines] if x!=['']]
    for i in range(len(xpath)):
        plt.plot(xpath[i], ypath[i], 'c-')
with open("path0.dat") as f1:
    lines = f1.read().split('\n')
    xpath = [ float(x[0]) for x in [y.split(',') for y in lines] if x!=['']]
    ypath = [ float(x[1]) for x in [y.split(',') for y in lines] if x!=['']]
    plt.plot(xpath, ypath, 'r-')
with open("path1.dat") as f1:
    lines = f1.read().split('\n')
    xpath = [ float(x[0]) for x in [y.split(',') for y in lines] if x!=['']]
    ypath = [ float(x[1]) for x in [y.split(',') for y in lines] if x!=['']]
    plt.plot(xpath, ypath, 'b-')
```
